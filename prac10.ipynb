```python
import pandas as pd
```

```python
df = pd.read_csv('/content/smartphone_usage_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#  Battery_Used_% by Daily_Screen_Time_Min (Bar plot)
plt.figure(figsize=(10, 6))
sns.barplot(x="Battery_Used_%", y="Daily_Screen_Time_Min", data=df, palette="Set2")
plt.title("Battery_Used_% by Daily_Screen_Time_Min (Bar plot)", fontsize=16)
plt.xlabel("Battery Used")
plt.ylabel("Daily Screen Time")
plt.tight_layout()
plt.show()
```

```python
#  Battery_Used_% by Daily_App_Opens (Bar plot)
plt.figure(figsize=(10, 6))
sns.barplot(x="Battery_Used_%", y="Daily_App_Opens", data=df, palette="Set2")
plt.title("Battery_Used_% by Daily_App_Opens (Bar plot)", fontsize=16)
plt.xlabel("Battery Used")
plt.ylabel("Daily App Opens")
plt.tight_layout()
plt.show()
```

```python
# Primary_App_Category vs Battery_Used_% (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x="Battery_Used_%", y="Primary_App_Category", data=df, palette="Set2")
plt.title("Primary App Category  vs Battery Used (Box Plot)", fontsize=16)
plt.ylabel("Primary App Category")
plt.xlabel("Battery Used")
plt.tight_layout()
plt.show()
```

```python
# heatmap correlation between numerical features
plt.figure(figsize=(10, 6))
correlation_matrix = df[['Daily_Screen_Time_Min',
'Daily_App_Opens','Notifications_Received',
'Battery_Used_%']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import axes
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# plot1: Battery Used vs Daily Screen Time
sns.barplot(x="Battery_Used_%", y="Daily_Screen_Time_Min",
            data=df, palette="Set2", ax=axes[0,0])
axes[0,0].set_title("Battery Used by Daily Screen Time", fontsize=12)

# plot2: Battery Used vs Daily App Opens
sns.barplot(x="Battery_Used_%", y="Daily_App_Opens",
            data=df, palette="Set2", ax=axes[0,1])
axes[0,1].set_title("Battery Used by Daily App Opens", fontsize=12)

# plot3: Primary App Category vs Battery Used
sns.boxplot(x="Battery_Used_%", y="Primary_App_Category",
            data=df, palette="Set2", ax=axes[1,0])
axes[1,0].set_title("Primary App Category vs Battery Used", fontsize=12)

# plot4: Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
            fmt=".2f", linewidths=0.5, ax=axes[1,1])
axes[1,1].set_title("Correlation Heatmap", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#Insights
insights = {
"Battery Used by Daily Screen Time":"Higher daily screen time leads to significantly higher battery consumption.",
"Battery Used by Daily App Opens":"Users who open apps more frequently uses more battery",
"Primary App Category by Battery Used ":"heavy app categories like video and gaming show the highest battery usage.",
"Correlation Heatmap":"Battery usage is most strongly correlated with screen time"
}
for key, value in insights.items():
  print(f"{key}: {value}\n")
```
