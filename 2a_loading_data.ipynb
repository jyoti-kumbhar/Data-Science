```python
#loading a sample dataset
from sklearn import datasets
digits = datasets.load_digits()
features = digits.data
targets = digits.target
features[0]
```

```python
#creating a simulated dataset(regression)
from sklearn.datasets import make_regression
features, target, coefficients = make_regression(n_samples = 200,n_features = 5,n_informative = 4,noise = 0.0,coef = True,random_state=1)
print("Feature Martix \n{}".format(features[:3]))
print("Target Vector \n{}".format(target[:3]))
```

```python
#creating a simulated dataset(classification)
from sklearn.datasets import make_classification
features, target = make_classification(n_samples = 200,n_features = 3,n_informative = 3,n_redundant = 0,n_classes = 2,weights=[.25,.75],random_state=1)
print("Feature Martix \n{}".format(features[:3]))
print("Target Vector \n{}".format(target[:3]))
```

```python
#creating a simulated dataset(clustering)
from sklearn.datasets import make_blobs
features, target = make_blobs(n_samples = 100,n_features = 2,centers = 3, cluster_std=0.5,shuffle = True,random_state=1)
print("Feature Martix \n{}".format(features[:3]))
print("Target Vector \n{}".format(target[:3]))

import matplotlib.pyplot as plt
plt.scatter(features[:,0],features[:,1],c=target)
plt.show()
```
