2. Conduct post-hoc tests to identify significant differences between group means.

```python
#ANOVA (F-TEST)
import pandas as pd
from scipy import stats
#hypothesis
H0='All education groups have the same default rate.'
H1='At least one education group has a different default rate.'
df = pd.read_csv('/content/Loan_default.csv')
# Corrected column names: 'Default' instead of 'Status', and using 'Education' as the grouping variable
df_anova = df[['Default', 'Education']]
# Drop missing values
df_anova = df_anova.dropna()
# Use 'Education' as the grouping variable for ANOVA
groups = df_anova['Education'].unique()
data = {grp: df_anova['Default'][df_anova['Education'] == grp] for grp in groups}
# Perform one-way ANOVA
F, p = stats.f_oneway(*[data[grp] for grp in groups])
print("F-statistic:", F)
print("p-value:", p)
if p < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

Conclusion:

The p-value is very small, showing that the groups are different. we reject the null hypothesis and the grouping variable has a real effect on the result.

```python
#Two Way F-test
import statsmodels.api as sm
from statsmodels.formula.api import ols
H0='Education level does not affect default rate'
H1='Education level does affect default rate.'
df_anova = df[['Default', 'Education', 'MaritalStatus']].dropna()
# Two-Way ANOVA model
model = ols('Default ~ C(Education) + C(MaritalStatus)', data=df_anova).fit()
# ANOVA Table
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
# Extract p-value for main effect of Education
p_value_education = anova_table.loc['C(Education)', 'PR(>F)']
print("\nP-value for Main Effect of Education:", p_value_education)
# Decision
if p_value_education < 0.05:
    print("Reject Null Hypothsis")
else:
    print("Accept the null hypothesis")
```

Conclusion:

The p-value is almost zero, which means default rates are different across education levels. So, education clearly affects loan default, and we reject the null hypothesis.
